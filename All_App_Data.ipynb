```python
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
first_20 = pd.read_csv('data/app_info.csv')
top_196 = pd.read_csv('data/top_196_grossing.csv')
top_paid = pd.read_csv('data/top_paid_games.csv')
grossing_games = pd.read_csv('data/grossing_apps.csv')
```

```python
files = [first_20, top_196, top_paid, grossing_games]
```

```python
df = pd.concat(files)
```

```python
df
```

```python
df = df.round({'score': 1})
```

```python
df[df['title'] == 'Wattpad - Read & Write Stories']
```

```python
df.at[353,"released"]="Mar 31, 2021"
```

```python
def bool_to_word(bool):
    if bool == True:
        return 'Free'
    else:
        return "Paid"
```

```python
df['free'] = df['free'].apply(lambda x: bool_to_word(x))
```

```python
df = df[['title', 'installs', 'score', 'reviews', 'price', 'size', 'genre', 'genreId', 'contentRating', 'released', 
         'version', 'androidVersionText', 'free']].reset_index(drop=True)
```

```python
df.rename(columns={'title': 'App', 'installs': 'Installs', 'score':'Rating', 'reviews': 'Reviews', 'price':'Price', 
                  'contentRating':'Content Rating', 'released':'Last Updated', 'version':'Current Ver', 
                   'androidVersionText':'Android Ver', 'size':'Size', 'genre':'Genres', 'genreId':'Category',
                  'free':'Type'}, inplace=True)
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], format="%b %d, %Y", errors='coerce')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Last Updated'])
release_month = df['Last Updated'].dt.month
df['Month'] = release_month
```

```python
df
```

```python
df.to_csv('new_app_data.csv', index=None, header=True)
```
