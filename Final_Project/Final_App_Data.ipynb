```python
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
games = pd.read_csv('data/google_games.csv')
apps_data = pd.read_csv('extract_data/apps_info.csv')
```

```python
games
```

```python
files = [games, apps_data]
```

```python
df = pd.concat(files)
```

```python
df
```

```python
df.info()
```

```python
df = df.round({'score': 1})
```

```python
def bool_to_word(bool):
    if bool == True:
        return 'Free'
    else:
        return "Paid"
```

```python
df['free'] = df['free'].apply(lambda x: bool_to_word(x))
```

```python
import datetime 

datetime.datetime.fromtimestamp(1510762834).strftime('%c')
```

```python
df['updated'] = pd.to_datetime(df['updated'], unit='s')
```

```python
df['date'] = df['updated'].dt.date
```

```python
df = df[['title', 'installs', 'score', 'reviews', 'price', 'size', 'genre', 'genreId', 'contentRating', 'date', 
         'version', 'androidVersionText', 'free']].reset_index(drop=True)
```

```python
df.rename(columns={'title': 'App', 'installs': 'Installs', 'score':'Rating', 'reviews': 'Reviews', 'price':'Price', 
                  'contentRating':'Content Rating', 'date':'Last Updated', 'version':'Current Ver', 
                   'androidVersionText':'Android Ver', 'size':'Size', 'genre':'Genres', 'genreId':'Category',
                  'free':'Type'}, inplace=True)
```

```python
df.head()
```

```python
# df['Last Updated'] = pd.to_datetime(df['Last Updated'], format="%B %d, %Y", errors='coerce')
df['lastupdate'] = (df['Last Updated'] -  df['Last Updated'].max()).dt.days
```

```python
df.head()
```

```python
df = df.drop(['Last Updated'], axis =1)
```

```python
df.to_csv('apps_data.csv', index=None, header=True)
```
